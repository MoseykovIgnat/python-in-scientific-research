```
#Func 1: f(x) = sin(x); x from -2Pi to 2Pi
#Fucn 2: f(x) = log(x); x from 1 to 5
from numpy import (linspace,pi,sin,gradient,log)
import matplotlib.pyplot as plt
from math import pi
```

```
def plot(func, a, b):
    """x и y для графика функции"""
    x = linspace(a,b,100)
    y = func(x)
    '''
    Производная для функции
    Таким образом, dydx будет вычисляться с использованием центральных разностей 
    и будет иметь ту же длину, что и y, в отличие от numpy.diff,
    который использует прямые разности и возвращает вектор размера (n-1).
    '''
    dx = x[1]-x[0]
    dydx = gradient(y, dx)

    print(func.__name__)
    fig, ax = plt.subplots()
    ax.plot(x, y, label=func.__name__)
    ax.plot( x, dydx, label = 'derivative of the ' + func.__name__)
    ax.grid()
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()
    fig.set_figheight(5)
    fig.set_figwidth(8)
```

```
plot(sin,-2*pi, 2*pi)
```

```
plot(log,1,5)
```
